```python
#Importing
import pandas as pd
import numpy as np
from array import *
import matplotlib.pyplot as plt

# Reading in the high school admissions data
admissions = pd.read_csv('AdmissionsCounties2018.csv')
np_admissions = np.array(admissions)
```

```python
# Reading in number of students per county
students_county = pd.read_csv('NumStudentsCounty.csv')
# put into np array
np_students_county = np.array(students_county)
```

```python
# printing admissions df
# admissions
# print students/county df
# students_county
```

```python
# print just county and students for 2018 - 2019
# alameda_students = students_county.loc[students_county['County'] == 'Alameda']
# alameda_students
students18 = students_county[['County','2018-19']]
# students18
```

```python
# Adding Admissions
temp = admissions.loc[admissions['Count'] == 'Adm']
```

```python
# Accessing County Names
county = temp.groupby(['County/State/ Territory','Count']).sum().reset_index()
# county
```

```python
# just the county and adm
county_adm = county[['County/State/ Territory','All']]
# county_adm
```

```python
# rename column
dict = {'County/State/ Territory': 'County', 'All': 'Adm'}
county_adm.rename(columns=dict,inplace=True)
# county_adm
```

```python
# combine 2 df (concat)
# df = pd.concat([county_adm, students18], axis = 1)
# df
```

```python
# combine 2 df (merge)

merged_county = county_adm.merge(students18)
# merged_county
```

```python
# 2018-19 into int from str
#merged_county['2018-19'] = pd.to_numeric(merged_county['2018-19'])
```

```python
# make adm and 2018-19 into a percent by dividing
percent = (merged_county['Adm']/merged_county['2018-19'])*100
# percent
```

```python
# combine percent into df
merged_percent = pd.concat([merged_county, percent], axis = 1)
merged_percent
```

```python
#rename percent column
dict = {0: 'Acc Rate'}
merged_percent.rename(columns=dict,inplace=True)
# np_merged_percent = np.array(merged_percent)
merged_percent
```

```python
# sorting by acc rates
sort_percent = merged_percent.sort_values('Acc Rate')
sort_percent = sort_percent.drop(39)
sort_percent
```

```python
np_merged_percent = np.array(sort_percent)
# county array for plot
np_county = np_merged_percent[:,0]
# percent array for plot
np_percent = np_merged_percent[:,3]
```

```python
# plotting lowest counties percent

# sizing plot
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)

# labels
titlelow = 'Acceptance Rates into UCs from Public High Schools in Each County (Lowest 15).'
xlab = 'County'
ylab = 'Acceptance Rate (%)'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(titlelow)

plt.plot(np_county[:15], np_percent[:15])
```

```python
# plotting highest counties percent

# sizing plot
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)

# labels
titlehi = 'Acceptance Rates into UCs from Public High Schools in Each County (Highest 15).'
xlab = 'County'
ylab = 'Acceptance Rate (%)'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(titlehi)

# axis = plt.gca()
# axis.set_xticklabels(axis.get_xticks(), rotation = 45)
# plt.draw()

plt.plot(np_county[34:], np_percent[34:])
```
